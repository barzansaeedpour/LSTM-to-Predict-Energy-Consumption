```python
# building input variable
def create_dataset(dataset, time_step = 1):
  dataX, dataY = [] , []
  for i in range(len(dataset)-time_step-1):
    a = dataset[i:(i+time_step),0]
    dataX.append(a)
    dataY.append(dataset[i + time_step,0])
  return np.array(dataX), np.array(dataY)
```

```python
time_step = 100
X_train, y_train = create_dataset(train_data, time_step)
X_test, ytest = create_dataset(test_data, time_step)
X_val, yval = create_dataset(val_data, time_step)
```

## Model Structure 3

For a more advanced LSTM-based network, you could introduce more sophisticated layers, such as Attention Mechanisms, and incorporate techniques like Residual Connections (for deeper models), and Temporal Convolutional Networks (TCN). These additions improve the model's ability to capture long-term dependencies, focus on important time steps, and reduce the likelihood of vanishing gradient problems.

Here’s a more advanced architecture that includes:

- **Attention Mechanism**: Helps the model focus on the most relevant parts of the input sequence.
- **Residual Connections**: A technique used in deep networks to allow gradients to flow more easily through the network.
- **Temporal Convolutional Network (TCN)**: For temporal patterns that LSTMs might miss.
- **Regularization**: Using Dropout and Batch Normalization to reduce overfitting.

Key Features of This Architecture:
- Attention Mechanism: Custom attention layer helps the model focus on important parts of the input sequence, improving the accuracy in sequence-to-sequence tasks.
- Residual Connections: These connections help when you add more layers, reducing the risk of vanishing gradients.
- Bidirectional LSTM: Captures temporal dependencies in both directions, enhancing the sequence learning.
- Dropout and Batch Normalization: Regularizes the model and stabilizes training.
- Learning Rate Scheduling: (Optional) Use a learning rate scheduler for better convergence.

Next Steps:
- Tune hyperparameters: You can experiment with the number of units in LSTM, dropout rates, and other hyperparameters to optimize performance.
- Further Extensions: Consider using a Temporal Convolutional Network (TCN) or incorporating Transformers if you are working with large datasets and need more powerful sequential modeling.

```python
from keras.models import Model
from keras.layers import Input, LSTM, Dense, Dropout, BatchNormalization, Bidirectional, Add, Layer
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping, ModelCheckpoint
import tensorflow as tf

# Custom attention layer
class AttentionLayer(Layer):
    def __init__(self):
        super(AttentionLayer, self).__init__()

    def build(self, input_shape):
        self.W = self.add_weight(name="attention_weight", shape=(input_shape[-1], input_shape[-1]),
                                 initializer="glorot_uniform", trainable=True)
        self.b = self.add_weight(name="attention_bias", shape=(input_shape[-1],), initializer="zeros", trainable=True)
        self.u = self.add_weight(name="context_vector", shape=(input_shape[-1], 1),
                                 initializer="glorot_uniform", trainable=True)

    def call(self, x):
        score = tf.nn.tanh(tf.tensordot(x, self.W, axes=1) + self.b)
        attention_weights = tf.nn.softmax(tf.tensordot(score, self.u, axes=1), axis=1)
        context_vector = attention_weights * x
        context_vector = tf.reduce_sum(context_vector, axis=1)
        return context_vector

# Model architecture
time_step = 100
input_seq = Input(shape=(time_step, 1))

# First Bidirectional LSTM Layer
x = Bidirectional(LSTM(100, return_sequences=True))(input_seq)
x = Dropout(0.3)(x)
x = BatchNormalization()(x)

# Second Bidirectional LSTM Layer with Residual Connection
x_res = Bidirectional(LSTM(100, return_sequences=True))(x)
x_res = Dropout(0.3)(x_res)
x_res = BatchNormalization()(x_res)
x = Add()([x, x_res])  # Adding residual connection

# Third LSTM Layer
x = Bidirectional(LSTM(100, return_sequences=True))(x)
x = Dropout(0.3)(x)
x = BatchNormalization()(x)

# Attention Layer
x = AttentionLayer()(x)

# Output layer
output = Dense(1)(x)

# Define the model
model = Model(inputs=input_seq, outputs=output)

# Compile the model
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mean_squared_error')

# Early stopping and model checkpointing
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = ModelCheckpoint('advanced_model.keras', monitor='val_loss', save_best_only=True)

# Summary of the model
model.summary()
```
